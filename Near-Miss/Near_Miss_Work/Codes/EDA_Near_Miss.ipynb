```python
import pandas as pd
import numpy as np
import plotly.express as px
import sweetviz as sv
from ydata_profiling import ProfileReport
from pathlib import Path

# ==========================
# LOAD DATA
# ==========================

file="merged_near_miss.xlsx"

df=pd.read_excel(
    file,
    dtype=str
)

print("Shape:",df.shape)

print("\nColumns:")
print(list(df.columns))

# ==========================
# STANDARDIZE EMPTY
# ==========================

df=df.replace(
    [
        "",
        "NA",
        "N/A",
        "nan"
    ],
    np.nan
)

# ==========================
# BASIC REPORT
# ==========================

print("\nMissing Values:")
print(
    df.isna().sum()
)

print("\nDuplicates:")
print(
    df.duplicated().sum()
)

# ==========================
# SWEETVIZ HTML
# ==========================

print("\nGenerating Sweetviz...")

sweet_report=sv.analyze(
    df
)

sweet_report.show_html(
    "near_miss_sweetviz_report.html"
)

print(
    "Saved: near_miss_sweetviz_report.html"
)

# ==========================
# YDATA PROFILE
# ==========================

print("\nGenerating Profile Report...")

profile=ProfileReport(

    df,

    title="Near Miss EDA",

    explorative=True

)

profile.to_file(
    "near_miss_profile_report.html"
)

print(
    "Saved: near_miss_profile_report.html"
)

# ==========================
# PLOTLY DASHBOARD
# ==========================

html=[]

html.append(
"""
<html>
<head>
<title>Near Miss EDA</title>
</head>
<body>
<h1>Near Miss EDA Dashboard</h1>
"""
)

# ==========================
# MISSING %
# ==========================

missing=(
    df
    .isna()
    .mean()
    *100
)

missing=(
    missing[
        missing>0
    ]
)

if len(missing)>0:

    fig=px.bar(

        x=missing.index,

        y=missing.values,

        title="Missing Percentage"

    )

    html.append(
        fig.to_html(
            full_html=False
        )
    )

# ==========================
# COLUMN DISTRIBUTIONS
# ==========================

cat=df.select_dtypes(
    include="object"
)

for col in cat.columns[:10]:

    try:

        temp=(
            df[col]
            .value_counts()
            .head(15)
        )

        if len(temp)>0:

            fig=px.bar(

                x=temp.index,

                y=temp.values,

                title=col

            )

            html.append(

                fig.to_html(
                    full_html=False
                )

            )

    except:
        pass

# ==========================
# UNIQUE COUNT
# ==========================

unique=(
    df
    .nunique()
)

fig=px.bar(

    x=unique.index,

    y=unique.values,

    title="Unique Values"

)

html.append(
    fig.to_html(
        full_html=False
    )
)

html.append(
"""
</body>
</html>
"""
)

with open(
    "near_miss_dashboard.html",
    "w",
    encoding="utf-8"
) as f:

    f.write(
        "\n".join(
            html
        )
    )

print(
    "\nSaved: near_miss_dashboard.html"
)

# ==========================
# SUMMARY
# ==========================

print("\nGenerated Files:")

print(
"""
1. near_miss_sweetviz_report.html
2. near_miss_profile_report.html
3. near_miss_dashboard.html
"""
)
```
