```python
import pandas as pd
import numpy as np
import sweetviz as sv
from ydata_profiling import ProfileReport
import plotly.express as px

# ==========================
# LOAD DATA
# ==========================

file="LARP_Database_Merged.xlsx"

df=pd.read_excel(
    file,
    dtype=str
)

print("Shape:",df.shape)

print("\nColumns:")
print(list(df.columns))

# ==========================
# STANDARDIZE EMPTY
# ==========================

df=df.replace(
    [
        "",
        "NA",
        "N/A",
        "nan",
        "NULL"
    ],
    np.nan
)

# ==========================
# BASIC SUMMARY
# ==========================

print("\nMissing Values:")
print(
    df.isna().sum()
)

print("\nDuplicates:")
print(
    df.duplicated().sum()
)

print("\nUnique Values:")
print(
    df.nunique()
)

# ==========================
# SWEETVIZ REPORT
# ==========================

print("\nGenerating Sweetviz Report...")

report=sv.analyze(
    df
)

report.show_html(
    filepath="larp_sweetviz_report.html",
    open_browser=False
)

print(
    "Saved: larp_sweetviz_report.html"
)

# ==========================
# PROFILE REPORT
# ==========================

print("\nGenerating Profile Report...")

profile=ProfileReport(

    df,

    title="LARP EDA Report",

    explorative=True,

    minimal=False

)

profile.to_file(
    "larp_profile_report.html"
)

print(
    "Saved: larp_profile_report.html"
)

# ==========================
# CUSTOM DASHBOARD
# ==========================

html=[]

html.append("""
<html>
<head>
<title>LARP Dashboard</title>
</head>
<body>
<h1>LARP EDA Dashboard</h1>
""")

# ==========================
# MISSING %
# ==========================

missing=(
    df
    .isna()
    .mean()
    *100
)

missing=(
    missing[
        missing>0
    ]
)

if len(missing)>0:

    fig=px.bar(

        x=missing.index,

        y=missing.values,

        title="Missing Percentage"

    )

    html.append(
        fig.to_html(
            full_html=False
        )
    )

# ==========================
# TOP CATEGORIES
# ==========================

for col in df.columns:

    try:

        temp=(
            df[col]
            .dropna()
            .value_counts()
            .head(15)
        )

        if (
            len(temp)>1
            and
            temp.shape[0]<=15
        ):

            fig=px.bar(

                x=temp.index,

                y=temp.values,

                title=col

            )

            html.append(
                fig.to_html(
                    full_html=False
                )
            )

    except:
        pass

# ==========================
# UNIQUE VALUES
# ==========================

unique=df.nunique()

fig=px.bar(

    x=unique.index,

    y=unique.values,

    title="Unique Values"

)

html.append(
    fig.to_html(
        full_html=False
    )
)

html.append(
"""
</body>
</html>
"""
)

with open(
    "larp_dashboard.html",
    "w",
    encoding="utf-8"
) as f:

    f.write(
        "\n".join(
            html
        )
    )

print(
    "Saved: larp_dashboard.html"
)

# ==========================
# DONE
# ==========================

print("\nGenerated Files:")

print(
"""
1. larp_sweetviz_report.html
2. larp_profile_report.html
3. larp_dashboard.html
"""
)
```
