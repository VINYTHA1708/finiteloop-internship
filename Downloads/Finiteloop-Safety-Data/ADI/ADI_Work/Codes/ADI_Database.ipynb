```python
import pandas as pd
import numpy as np

# ==========================
# LOAD DATA
# ==========================

file="../ADIDatabase.xlsx"

df=pd.read_excel(
    file,
    dtype=str
)

print("Original Shape:",df.shape)

# ==========================
# CLEAN COLUMN NAMES
# ==========================

df.columns=(
    df.columns
    .astype(str)
    .str.strip()
    .str.replace("\n"," ",regex=False)
    .str.replace(r"\s+","_",regex=True)
)

# ==========================
# STANDARDIZE EMPTY VALUES
# ==========================

empty_values=[

    "",
    "NA",
    "N/A",
    "na",
    "n/a",
    "NULL",
    "null",
    "nan"

]

df=df.replace(
    empty_values,
    np.nan
)

# ==========================
# REMOVE 100% EMPTY COLUMNS
# ==========================

drop_cols=[]

for col in df.columns:

    empty=(

        df[col]
        .fillna("")
        .astype(str)
        .str.strip()
        .eq("")
        .all()

    )

    if empty:

        drop_cols.append(col)

df=df.drop(
    columns=drop_cols,
    errors="ignore"
)

print("\nRemoved Empty Columns:")
print(drop_cols)

# ==========================
# REMOVE DUPLICATES
# ==========================

duplicates=df.duplicated().sum()

df=df.drop_duplicates()

print("\nDuplicates Removed:",duplicates)

# ==========================
# RESET INDEX
# ==========================

df=df.reset_index(
    drop=True
)

# ==========================
# ADD SI_No
# ==========================

df.insert(
    0,
    "SI_No",
    np.arange(
        1,
        len(df)+1
    )
)

print("\nSI_No added and reordered")

# ==========================
# MISSING SUMMARY
# ==========================

missing=(

    df
    .replace("",np.nan)
    .isna()
    .sum()

)

missing=missing[
    missing>0
]

print("\nMissing Values Summary:")
print(missing)

# ==========================
# FINAL REPORT
# ==========================

print("\nFinal Shape:")
print(df.shape)

print("\nColumns After Cleaning:")
print(list(df.columns))

# ==========================
# EXPORT
# ==========================

output="cleaned_ADIDatabase.xlsx"

with pd.ExcelWriter(
    output,
    engine="openpyxl"
) as writer:

    df.to_excel(
        writer,
        index=False,
        sheet_name="Cleaned_Data"
    )

print("\nSaved:",output)
```
